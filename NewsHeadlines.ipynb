```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and preprocess the data
df = pd.read_csv("/content/archive.zip")
df['date'] = pd.to_datetime(df['publish_date'], format='%Y%m%d')
df['year'] = df['date'].dt.year

# Plot headlines per year
plt.figure(figsize=(10, 5))
df['year'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Headlines Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Extract month from date
df['month'] = df['date'].dt.month

# Plot headlines per month
plt.figure(figsize=(10, 5))
df['month'].value_counts().sort_index().plot(kind='bar', color='lightgreen')
plt.title('Total Headlines Per Month (Across All Years)')
plt.xlabel('Month')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import re

# Combine all headlines into a single string
all_words = ' '.join(df['headline_text'].dropna().tolist()).lower()
words = re.findall(r'\b\w+\b', all_words)

# Count top 20 words
common_words = Counter(words).most_common(20)
labels, values = zip(*common_words)

# Plot word frequency
plt.figure(figsize=(12, 6))
plt.bar(labels, values, color='coral')
plt.title('Top 20 Most Common Words in Headlines')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group headlines by month and count
monthly_counts = df.groupby(df['date'].dt.to_period('M')).size()

# Plot
plt.figure(figsize=(14, 6))
monthly_counts.plot(kind='line', color='purple')
plt.title('Monthly Trend of ABC News Headlines')
plt.xlabel('Month')
plt.ylabel('Number of Headlines')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
top_years = df['year'].value_counts().nlargest(5)

# Plot
plt.figure(figsize=(7, 7))
plt.pie(top_years.values, labels=top_years.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Proportion of Headlines from Top 5 Years')
plt.tight_layout()
plt.show()
```

```python
# Filter for a specific year
df_2005 = df[df['year'] == 2005]
daily_counts = df_2005.groupby(df_2005['date'].dt.date).size()

# Plot
plt.figure(figsize=(14, 6))
plt.plot(daily_counts.index, daily_counts.values, color='orange')
plt.title('Number of Headlines Per Day in 2005')
plt.xlabel('Date')
plt.ylabel('Headlines')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Calculate headline lengths in number of words
headline_lengths = df['headline_text'].dropna().apply(lambda x: len(x.split()))

# Plot
plt.figure(figsize=(10, 5))
plt.hist(headline_lengths, bins=30, color='steelblue', edgecolor='black')
plt.title('Distribution of Headline Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/archive.zip")

# Convert to datetime
df['date'] = pd.to_datetime(df['publish_date'], format='%Y%m%d')

# Sort by date (ascending)
df_sorted_by_date = df.sort_values(by='date', ascending=True)

# Display first 10 sorted rows
print(df_sorted_by_date.head(10))
```

```python
# Sort by headline_text alphabetically
df_sorted_by_headline = df.sort_values(by='headline_text', ascending=True)

# Display first 10 rows
print(df_sorted_by_headline.head(10))
```

```python
# Add a new column for headline length
df['headline_length'] = df['headline_text'].apply(lambda x: len(str(x).split()))

# Sort by headline length (descending)
df_sorted_by_length = df.sort_values(by='headline_length', ascending=False)

# Display top 10 longest headlines
print(df_sorted_by_length.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Convert 'publish_date' to datetime and extract year and month
df['date'] = pd.to_datetime(df['publish_date'], format='%Y%m%d')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month


# Create a pivot table of year vs month
pivot = df.pivot_table(index='year', columns='month', values='headline_text', aggfunc='count').fillna(0)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot, cmap='YlGnBu', annot=True, fmt=".0f")
plt.title('Heatmap of Headlines by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```

```python
keyword = "election"

# Filter headlines containing the keyword
df_keyword = df[df['headline_text'].str.contains(keyword, case=False, na=False)]

# Group by year
keyword_trend = df_keyword['year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
plt.plot(keyword_trend.index, keyword_trend.values, marker='o', color='crimson')
plt.title(f'Usage Trend of "{keyword}" in Headlines Over Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Ensure headline_length column exists
df['headline_length'] = df['headline_text'].apply(lambda x: len(str(x).split()))

# Plot boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(x='year', y='headline_length', data=df[df['year'] >= 2000])
plt.title('Distribution of Headline Lengths per Year (2000 Onwards)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_10_years = df['year'].value_counts().nlargest(10)

# Plot
plt.figure(figsize=(10, 5))
top_10_years.sort_index().plot(kind='bar', color='darkcyan')
plt.title('Top 10 Years with Most Headlines')
plt.xlabel('Year')
plt.ylabel('Number of Headlines')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
daily_counts = df['date'].value_counts().sort_index()
rolling_avg = daily_counts.rolling(window=7).mean()

plt.figure(figsize=(14, 6))
plt.plot(daily_counts.index, daily_counts.values, alpha=0.3, label='Daily Count')
plt.plot(rolling_avg.index, rolling_avg.values, color='blue', label='7-Day Rolling Avg')
plt.title('Rolling Average of Daily Headlines')
plt.xlabel('Date')
plt.ylabel('Headline Count')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group by date
top_days = df['date'].value_counts().nlargest(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=top_days.index.strftime('%Y-%m-%d'), y=top_days.values, palette='Reds')
plt.title('Top 10 Days with Most Headlines')
plt.xticks(rotation=45)
plt.ylabel('Number of Headlines')
plt.tight_layout()
plt.show()
```

```python
df['weekday'] = df['date'].dt.day_name()
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(10, 5))
sns.countplot(x='weekday', data=df, order=weekday_order, palette='Set2')
plt.title('Headline Count by Day of the Week')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Vectorize to get bigrams
vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
X = vectorizer.fit_transform(df['headline_text'].dropna())

# Sum frequencies
bigrams_freq = X.sum(axis=0).A1
bigrams = vectorizer.get_feature_names_out()
freq_df = pd.DataFrame({'bigram': bigrams, 'freq': bigrams_freq})
top_bigrams = freq_df.sort_values(by='freq', ascending=False).head(20)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='freq', y='bigram', data=top_bigrams, palette='magma')
plt.title('Top 20 Most Frequent Bigrams in Headlines')
plt.xlabel('Frequency')
plt.ylabel('Bigram')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(df['headline_length'], shade=True, color='purple')
plt.title('Kernel Density Estimate of Headline Lengths')
plt.xlabel('Number of Words')
plt.tight_layout()
plt.show()
```
